```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df["avg_score"] = df[["math score","reading score","writing score"]].mean(axis=1)
```

BAR PLOT

```python
df.groupby("gender")["avg_score"].mean().plot(kind="bar")
plt.title("Average Score by Gender")
plt.show()
```

PIE CHART

```python
df["test preparation course"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Test Preparation Distribution")
plt.ylabel("")
plt.show()
```

SCATTER PLOT

```python
plt.scatter(df["math score"], df["reading score"])
plt.xlabel("Math")
plt.ylabel("Reading")
plt.show()
```

Histogram (Score Distribution)

```python
plt.hist(df["avg_score"], bins=10)
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Average Scores")
plt.show()
```

VIOLIN PLOT

```python
sns.violinplot(x="gender", y="avg_score", data=df)
plt.title("Score Distribution by Gender")
plt.show()
```

Line Plot (Trend-like Visualization)

```python
df.sort_values("avg_score").reset_index(drop=True)["avg_score"].plot()
plt.title("Sorted Average Scores")
plt.ylabel("Score")
plt.show()
```

HEATMAPS

```python
sns.heatmap(df[["math score","reading score","writing score"]].corr(), annot=True)
plt.show()
```

CORELLOGRAM

```python
corr = df[["math score", "reading score", "writing score"]].corr()

plt.figure(figsize=(6,5))
sns.scatterplot(
    x=corr.columns.repeat(len(corr)),
    y=corr.index.tolist() * len(corr),
    size=corr.values.flatten(),
    hue=corr.values.flatten(),
    sizes=(100, 1000),
    palette="coolwarm",
    legend=False
)

plt.title("Correlogram of Subject Scores")
plt.xticks(rotation=45)
plt.show()
```
